```python
import pandas as pd
import numpy as np

#path_prefix = "/home/jovyan/work/data/"
path_prefix = "/data/prod"
```

```python
portal = pd.read_csv('../../../../data/prod/treas_parking_payments_2017_datasd.csv')
portal.head()
```

```python
portal['trans_start'] = pd.to_datetime(portal['trans_start'],  format="%Y-%m-%d %H:%M:%S")
portal['month'] = portal.trans_start.dt.strftime("%m")
portal['day'] = portal.trans_start.dt.strftime("%d")
portal.head()
```

```python
portal['month'] = portal.trans_start.dt.month
portal['day'] = portal.trans_start.dt.day
portal.head()
```

```python
grouped = portal.groupby(['pole_id', 'month'], as_index=False)
by_month = grouped['trans_amt'].agg({'sum_trans_amt': 'sum', 'num_trans': 'count'})

by_month.head()
```

```python
grouped = portal.groupby(['pole_id', 'month', 'day'], as_index=False)
by_day = grouped['trans_amt'].agg({'sum_trans_amt': 'sum', 'num_trans': 'count'})

by_day.head()
```

```python
by_month.to_csv('../../../../data/prod/treas_meters_2017_pole_by_month_datasd.csv', index=True, encoding='utf-8', doublequote=True, date_format="%Y-%m-%d %H:%M:%S")
by_day.to_csv('../../../../data/prod/treas_meters_2017_pole_by_mo_day_datasd.csv', index=True, encoding='utf-8', doublequote=True, date_format="%Y-%m-%d %H:%M:%S")
```
