```python
cf2 = cf[[
    'CampusGuid',
    'Year',
    'Condition',
    'CapitalRequirement',
    'MaintenanceRequirement',
    'TotalRequirement',
    'FacilityConditionIndex'
]].rename(columns={
    "CampusGuid": "campus_guid",
    "Year": "survey_year",
    "Condition": "condition",
    "FacilityConditionIndex": "fci",
    "CapitalRequirement": "capital_req",
    "MaintenanceRequirement": "maintenance_req",
    "TotalRequirement": "total_req"
})
cf2 = cf2[cf2.survey_year == 2015].copy().reset_index().drop(['index'], axis = 1)
cf2.head()
```
